```python
import folium
from folium.plugins import HeatMap

def plot_heatmap(df):
  # Initialize map centered around the first data point
  m = folium.Map(location=[df.iloc[0]['Latitude'], df.iloc[0]['Longitude']], zoom_start=12)

  # Create a list of coordinates for the heatmap
  heat_data = [[row['Latitude'], row['Longitude']] for index, row in df.iterrows()]

  # Add HeatMap to the map
  HeatMap(heat_data).add_to(m)

  # Display the map (if in a Jupyter notebook environment)
  return m
```
